```python
import torch
from diffusers import StableDiffusionPipeline
import os

# 1. إعداد المحرك (نحتاج تشغيله مرة واحدة)
device = "cuda" if torch.cuda.is_available() else "cpu"
model_id = "runwayml/stable-diffusion-v1-5"
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16 if device == "cuda" else torch.float32)
pipe = pipe.to(device)

def generate_studio_shot(key):
    # الذاكرة المدمجة: دمج الخيال العلمي مع الاستوديو
    preference = "Sci-Fi futuristic studio background, subtle holographic screens, soft cinematic lighting"

    # صياغة النص بناءً على وصفك الدقيق
    user_prompt = "A person, casually dressed, sitting on a studio chair. Low-angle shot from the extreme bottom-left, looking upwards. The person is speaking and gesturing, looking to the right (not at the camera). Large professional cameras visible in the distance. Futuristic lighting. Realistic photo."
    final_prompt = f"{user_prompt}, {preference}, high quality, 8k, detailed faces, photorealistic."

    print(f"--- تنفيذ الأمر ---")
    print(f"الأمر النهائي: {final_prompt}")

    image = pipe(final_prompt).images[0]
    filename = f"studio_shot_{key}.png"
    image.save(filename)
    print(f"✅ تم حفظ الصورة باسم {filename}!")

# تجربة التشغيل
generate_studio_shot(1)
```
